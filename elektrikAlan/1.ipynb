```python
import numpy as np
from scipy. optimize import bisect
#
def fun(x):
    return x**2 -3
#
a=1
b=3
adım_sayısı=1.0e-15
#
fa=fun (a)
fb=fun (b)
#
if fa*fb>0:
    print("wrong interval !!!" ,fa , fb)
    exit ( )
#
x=bisect (fun , a,b, xtol=adım_sayısı )

print(x)
```

```python
def f_example(x):
    return x - np.exp(-x)


def secant_method(f, x0, x1, Deneme_Sayısı):
    for i in range(Deneme_Sayısı):
        payda = float(f(x1) - f(x0))
        if payda == 0:
            quit('')

        x2 = x1 - f(x1) * (x1 - x0) / payda
        x0, x1 = x1, x2
    return x2

root = secant_method(f_example, 10, 30, 8)

print(root)


root1 = secant_method(f_example, 10, 30, 77)

print(root1)


```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')


f = lambda x, y: np.exp(-x) 
delta_x = 0.1 
x = np.arange(0, 1 + delta_x, delta_x) 
y0 = -1 


y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + delta_x*f(x[i], y[i])

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo--', label='Yaklaşık')
plt.plot(x, -np.exp(-x), 'g', label='Doğrusu')
plt.title('e^{-x} ')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
