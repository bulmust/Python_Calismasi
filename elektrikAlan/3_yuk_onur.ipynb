```python
import matplotlib.pyplot as plt
import numpy as np
#axis ayar/ limits of the axis
N = 25
M = 25
#koordinatlar / coordinates and the grid
X = np.arange(0, M, 1)
Y = np.arange(0, N, 1)
X, Y = np.meshgrid(X, Y)

Ex ,Ey = np.zeros((N,M)), np.zeros((N,M)) #Elektrik alanmn x ve y si iin depo / a depo for my electric fields x and y component

konum = [[], []] #location of the particle

konum_x = 21 # x location
konum_y = 22 # y location

konum[0].append(konum_x)
konum[1].append(konum_y)

yuk = 11*10**(-9) # charge is something like + 11 coulomb

for i in range(N):
    for j in range(M):
        payda = ((i - konum_y)**2 + (j - konum_x)**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
        if payda != 0:
             Ex[i, j] += yuk*(8.99*10**9)*(j - konum_x) / payda
             Ey[i, j] += yuk*(8.99*10**9)*(i - konum_y) / payda




Ex1 ,Ey1 = np.zeros((N,M)), np.zeros((N,M))

konum1 = [[], []] #location of the particle

konum_x1 = 13 # x location
konum_y1 = 14 # y location

konum1[0].append(konum_x1)
konum1[1].append(konum_y1)

yuk1 = -15*10**(-9) # charge is something like + 11 coulomb

for i in range(N):
    for j in range(M):
        payda1 = ((i - konum_y1)**2 + (j - konum_x1)**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
        if payda1 != 0:
             Ex1[i, j] += yuk1*(8.99*10**9)*(j - konum_x1) / payda1
             Ey1[i, j] += yuk1*(8.99*10**9)*(i - konum_y1) / payda1



Ex2 ,Ey2 = np.zeros((N,M)), np.zeros((N,M))

konum2 = [[], []] #location of the particle

konum_x2 = 7 # x location
konum_y2 = 11 # y location

konum2[0].append(konum_x2)
konum2[1].append(konum_y2)

yuk2 = 21*10**(-9) # charge is something like + 11 coulomb

for i in range(N):
    for j in range(M):
        payda2 = ((i - konum_y2)**2 + (j - konum_x2)**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
        if payda2 != 0:
             Ex1[i, j] += yuk2*(8.99*10**9)*(j - konum_x2) / payda2
             Ey1[i, j] += yuk2*(8.99*10**9)*(i - konum_y2) / payda2

plt.figure(figsize=(12, 8))

# field
plt.plot(*konum, 'bo')
plt.plot(*konum1, 'bo')
plt.plot(*konum2, 'bo')
plt.quiver(X, Y, Ex, Ey, scale = 500, pivot='mid')
plt.quiver(X, Y, Ex1, Ey1, scale = 500, pivot='mid')
plt.quiver(X, Y, Ex2, Ey2, scale = 500, pivot='mid')
# colorbar for magnitude
# misc
plt.title('Elektirk Alan')
plt.axis('equal')
plt.axis()
plt.show()
```
