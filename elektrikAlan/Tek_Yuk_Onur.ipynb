```python
import matplotlib.pyplot as plt
import numpy as np
import random
#axis ayarı/ limits of the axis
N = 25
M = 25
#koordinatlar / coordinates and the grid
X = np.arange(0, M, 1)
Y = np.arange(0, N, 1)
X, Y = np.meshgrid(X, Y)

Ex ,Ey = np.zeros((N,M)), np.zeros((N,M)) #Elektrik alanımın x ve y si için depo / a depo for my electric fields x and y component

konum = [[], []] #location of the particle

konum_x = 12 # x location
konum_y = 12 # y location

konum[0].append(konum_x)
konum[1].append(konum_y)

yuk = 11*10**(-9) # charge is something like + 11 coulomb

for i in range(N):
    for j in range(M):
        payda = ((i - konum_y)**2 + (j - konum_x)**2)**1.5  #payda = denominator, denom of E. field equation is ((x-x_0)^2 + (y - y_0)^2)*\sqrt((x-x_0)^2 + (y - y_0)^2)
        if payda != 0:
             Ex[i, j] += yuk*(8.99*10**9)*(j - konum_x) / payda
             Ey[i, j] += yuk*(8.99*10**9)*(i - konum_y) / payda

plt.figure(figsize=(12, 8))

# field
plt.plot(*konum, 'bo')
plt.quiver(X, Y, Ex, Ey, scale = 100, pivot='mid')
# colorbar for magnitude
# misc
plt.title('Elektirk Alan')
plt.axis('equal')
plt.axis()
plt.show()
```
